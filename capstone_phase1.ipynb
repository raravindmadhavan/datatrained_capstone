```python
def scrape_allen_solly_clothing(gender, clothing_type,weartype):
    base_url = "https://allensolly.abfrl.in/c/"
    url = f"{base_url}{gender}-{clothing_type}"
    driver=webdriver.Chrome() # Opening Chrome
    driver.get(url) # Opens the website
    time.sleep(5)
    clothing_data = []
    driver.maximize_window()
    for i in driver.find_elements(By.XPATH,"//div[starts-with(@class,'ProductCard_imageContainer')]"):
        time.sleep(5)
        try:
            i.click()
        except:
            continue
        driver.switch_to.window(driver.window_handles[-1])
        response = requests.get(driver.current_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        product_name=driver.find_element(By.XPATH,'//*[contains(@class,"productTitle")]').text
        product_details = soup.find_all(class_='MuiAccordionDetails-root ProductDetails_details__xjHof css-u7qq7e')[0].text.strip()
        product_code = re.split('StyleCode:',product_details)[1][:14]
        brand=re.split('Brand:',product_details)[1][:11] 
        material=re.split('(?<=[a-z])(?=[A-Z])',re.split('Material:',product_details)[1])[0]
        occasion=re.split('Occasion:',product_details)[1][:6] 
        p_type=soup.find_all(class_='ProductDeclaration_detailsList__JgIAQ')[0].text.strip()
        price=soup.find_all(class_='price')[0].text.strip() 
        color=soup.find_all(class_='Color_pdpColor_text_value__CtGmk')[0].text.strip() 
        country=soup.find_all(class_='ProductDeclaration_detailsItem__vALdd')[3].text.strip()
        return_time=30
        sizes=driver.find_elements(By.XPATH,'//*[contains(@class,"Size_pdpSize_content_size")]')
        size_list=[]
        for size in sizes:
            size_list.append(size.text)
        clothing_data.append({"ProductName": product_name,"Material":material, "Price": price,"Brandname":brand,
                              "Category":gender,"Producttype":p_type,"Occasion":occasion,"color":color,
                              "ProductCode":product_code,"Country":country,"return":return_time,"Size":','.join(size_list),"Weartype":weartype})
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

    return clothing_data

cloth=(("women","shirts","Topwear"),("women","tops","Topwear"),
       ("women","tunics","Topwear"),("women","dresses","Topwear"),
       ("women","winterwear","Topwear"),("women","blazers","Topwear"),
       ("women","trousers","Bottomwear"),("women","jeans","Bottomwear"),
       ("women","skirts","Bottomwear"),("women","jumpsuits","Topwear"),
       ("women","shorts-capris","Bottomwear"),("women","leggings-jeggings","Bottomwear"),
       ("women","joggers","Bottomwear"),("women","track-pants","Bottomwear"),
       ("men","shirts","Topwear"),("men","t-shirts","Topwear"),("men","winterwear","Topwear"),
       ("men","jeans","Bottomwear"),("men","suits","Topwear"),("men","waistcoats","Topwear"),
       ("men","blazers","Topwear"),("men","trousers","Bottomwear"),("men","shorts","Bottomwear"),("men","track-pants-and-joggers","Bottomwear"),
       ("boys","shirts","Topwear"),("boys","polo-t-shirts","Topwear"),("boys","round-neck-and-v-neck-t-shirts","Topwear"),
       ("boys","sweatshirts","Topwear"),("boys","sweaters","Topwear"),("boys","waistcoats","Topwear"),
       ("boys","suits-and-blazers","Topwear"),("boys","jackets","Topwear"),("boys","shorts","Bottomwear"),
       ("boys","trousers","Bottomwear"),("boys","track-pants","Bottomwear"),("boys","jeans","Bottomwear"),
       ("girls","tops-and-t-shirts","Topwear"),("girls","shirts","Topwear"),("girls","frocks-dresses","Topwear"),
       ("girls","sweatshirts","Topwear"),("girls","sweaters","Topwear"),("girls","jackets","Topwear"),("girls","shorts","Bottomwear"),
       ("girls","trousers","Bottomwear"),("girls","track-pants","Bottomwear"),("girls","jeans","Bottomwear"),
       ("girls","skirts","Bottomwear"),("girls","leggings-and-jeggings","Bottomwear")
     )
data=[]
for cloth_tuple in cloth:
    gender=cloth_tuple[0]
    clothing_type=cloth_tuple[1]
    weartype=cloth_tuple[2]
    data.extend(scrape_allen_solly_clothing(gender,clothing_type,weartype))
# if data:
#     print(f"Found {len(data)} {gender} {clothing_type}:")
```

```python
def scrape_usp_clothing(gender, clothing_type,weartype):
    
    base_url = "https://uspoloassn.in"
    if gender=='men':
        url = f"{base_url}/collections/{gender}-{clothing_type}"
        if clothing_type in ('jeans','trousers'):
            url=url+'-1'
    elif gender=='women':
        url = f"{base_url}/collections/{gender}-clothing-{clothing_type}"
    elif gender=='boys' and clothing_type not in ('jeans','trousers'):
        url = f"{base_url}/collections/{gender}-{clothing_type}"
    elif gender=='boys' and clothing_type in ('jeans','trousers'):
        url = f"{base_url}/collections/kids-{gender}-{clothing_type}"
    driver=webdriver.Chrome() # Opening Chrome
    driver.get(url) # Opens the website
    time.sleep(5)
    clothing_data = []
    driver.maximize_window()
    for i in driver.find_elements(By.XPATH,"//img[contains(@class,'first-image')]"):
        try:
            i.click()
        except:
            print(1)
            continue
        response = requests.get(driver.current_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        product_name=driver.find_element(By.XPATH,'//*[contains(@class,"product__section-title")]').text
        product_code=driver.find_element(By.XPATH,'//*[contains(@class,"variant_sku")]').text
        price=driver.find_element(By.XPATH,'//*[contains(@class,"price-item--sale")]').text
        sizes=driver.find_elements(By.XPATH,'//div[contains(@class,"swatches__container size js-variant-selector")]/div[contains(@class,"swatches__swatch--regular swatch-element  js-swatch-element")]')
        size_list=[]
        for size in sizes:
            size_list.append(size.text.strip())
        colors=driver.find_elements(By.XPATH,'//*[contains(@class,"product-swatch-title-container")]/span[contains(@class,"swatch-type")]')
        color_list=[]
        for color in colors:
            color_list.append(color.text.strip())
        material=re.split(r'[|-]',soup.find_all(class_='c-accordion__inner')[2].text.strip())[1].strip()
        country=soup.find_all(class_='c-accordion__inner')[4].find('p').text.split("Country of Origin:")[-1]
        occasion='Casual'
        brand='US Polo Assn'
        p_type=clothing_type
        return_time=re.findall(r'\d+\s+days',soup.find_all(class_='c-accordion__inner')[3].text.strip())[0]
        clothing_data.append({"ProductName": product_name,"Material":material, "Price": price,"Brandname":brand,
                              "Category":gender,"Producttype":p_type,"Occasion":occasion,"color":','.join(color_list),
                              "ProductCode":product_code,"Country":country,"return":return_time,"Size":','.join(size_list),"Weartype":weartype})
        driver.back()

    return clothing_data

cloth_usp=(("men","polo-shirts","Topwear"),("men","shirts","Topwear"),("men","t-shirts","Topwear"),
          ("men","trousers","Bottomwear"),("men","jeans","Bottomwear"),("men","shorts","Bottomwear"),
          ("women","polo-shirts","Topwear"),("women","tops","Topwear"),("women","dresses","Topwear"),
          ("women","t-shirts","Topwear"),("women","shirts","Topwear"),("women","jeans","Bottomwear"),
          ("boys","polo-shirts","Topwear"),("boys","shirts","Topwear"),("boys","t-shirts","Topwear"),
          ("boys","shorts","Bottomwear"),("boys","jeans","Bottomwear"),("boys","trousers","Bottomwear"))
for cloth_tuple in cloth_usp:
    gender=cloth_tuple[0]
    clothing_type=cloth_tuple[1]
    weartype=cloth_tuple[2]
    data.extend(scrape_usp_clothing(gender,clothing_type,weartype))
```

```python
def scrape_manyavar_clothing(clothing_type,gender):
    
    base_url = "https://www.manyavar.com"
    if clothing_type=='Kurta Pajama' and gender=='Men':
        url = f"{base_url}/en-in/men/kurta-pajama"
    elif clothing_type=='Kurta Jacket Set' and gender=='Men':
        url = f"{base_url}/en-in/men/kurta-jacket-set"
    elif clothing_type=='Kurta'and gender=='Men':
        url = f"{base_url}/en-in/men/kurta"
    elif clothing_type=='Jacket'and gender=='Men':
        url = f"{base_url}/en-in/men/nehru-jacket"
    elif clothing_type=='Indo Western' and gender=='Men':
        url = f"{base_url}/en-in/men/indo-western"
    elif clothing_type=='Sherwani' and gender=='Men':
        url = f"{base_url}/en-in/men/sherwani"
    elif clothing_type=='Lehenga' and gender=='Women':    
        url=f"{base_url}/en-in/mohey/lehenga"
    elif clothing_type=='Indo Western' and gender=='Women':
        url = f"{base_url}/en-in/mohey/indo-western"
    weartype='Topwear'
    driver=webdriver.Chrome() # Opening Chrome
    driver.get(url) # Opens the website
    time.sleep(5)
    clothing_data = []
    driver.maximize_window()
    response = requests.get(driver.current_url)
    ind_soup = BeautifulSoup(response.text, 'html.parser')
    for i in ind_soup.find_all(class_='tile-body'):
        driver.get(base_url+i.find_all('div')[0].find('h4').find('a').get('href'))
        response = requests.get(driver.current_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        product_name=soup.find(class_='attributes').find('div').find('div').find('h1').get_text().strip()
        product_code=soup.find(class_='sku-text').get_text().strip().split('SKU ID-')[-1]
        price=soup.find(class_='attributes').find_all('div')[2].find('h2').get_text().strip()
        sizes=soup.find(class_='size-button').find_all('a')
        size_list=[]
        for size in sizes:
            if size.text.strip()!='':
                size_list.append(size.get_text().strip())
        colors=soup.find(lambda tag:'Color' in tag.text and tag.name=='li').get_text().strip().split('Color:')[-1].strip()
        material=soup.find(lambda tag:'Fabric' in tag.text and tag.name=='li').get_text().strip().split('Fabric:')[-1].strip()
        country='India'
        occasion='Festive'
        brand='Manyavar'
        p_type=clothing_type
        return_time=15
        clothing_data.append({"ProductName": product_name,"Material":material, "Price": price,"Brandname":brand,
                              "Category":gender,"Producttype":p_type,"Occasion":occasion,"color":colors,
                              "ProductCode":product_code,"Country":country,"return":return_time,"Size":','.join(size_list),"Weartype":weartype})
        print(clothing_data)
    return clothing_data

for clothing_type,gender in (('Kurta Pajama','Men'),('Kurta Jacket Set','Men'),('Jacket','Men'),('Kurta','Men'),
                             ('Indo Western','Men'),('Sherwani','Men'),('Lehenga','Women'),('Indo Western','Women')):
    data.extend(scrape_manyavar_clothing(clothing_type,gender))
```
